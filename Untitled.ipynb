```python
%matplotlib inline

import pandas as pd
import numpy as np
import scipy.stats as scs
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

### Example

    Let X denote the IQ of a randomly selected adult American. Assume, a bit unrealistically, that X is normally distributed with unknown mean μ and standard deviation 16. Take a random sample of n = 16 students, so that, after setting the probability of committing a Type I error at α = 0.05, we can test the null hypothesis H0: μ = 100 against the alternative hypothesis that HA: μ > 100.

    What is the power of the hypothesis test if the true population mean were μ = 108?

```python
from math import sqrt

miu_hat = 100 #null hypothesis mean
miu = 108     #true population mean
sd  = 16      #standard deviation of the population

x = np.linspace(0, 200, 600)                         #generating x values to plug into the pdf function
pdf_y_pop = scs.norm(miu, sd/sqrt(16)).pdf(x)        #getting the pdf values for population mean
pdf_y_sample = scs.norm(miu_hat, sd/sqrt(16)).pdf(x) #getting the pdf for sample mean

alpha_value = scs.norm.ppf(0.95, miu_hat, sd/sqrt(16)) #calculating the  value for rejection threshold. based on alpha 
                                                       #level = 0.95

# finding the area to the left of this value based on sampling distribution 
area_left_sample = scs.norm.cdf(alpha_value, miu, sd/sqrt(16))
power = 1 - area_left_sample
print alpha_value
print area_left_sample
print power
```

```python
fig, ax = plt.subplots(figsize=(16,6))
_= ax.plot(x, pdf_y_sample, 
          color = 'b',
          label = '$H_0: $Null Hypothesis')
_= ax.plot(x, pdf_y_pop, 
           color = 'r',
          label = 'Sample Distribution')
_= ax.plot([miu, miu],[0,0.1], 
          color = 'r',
          linestyle = ':')
_= ax.plot([miu_hat, miu_hat],[0,0.1], 
          color = 'b',
          linestyle = ':')
_= ax.set_xlim(85, 125)
_= ax.set_ylim(0, 0.12)
_= ax.fill_between(x, pdf_y_sample, 0, 
                where=( (alpha_value <= x) ), 
                color="blue", 
                alpha=0.4,
                label = '$H_0$ Rejection Area')
_= ax.fill_between(x, pdf_y_pop, 0, 
                where=( (alpha_value <= x) ), 
                color="red", 
                alpha=0.3,
                label = 'Power')
_= ax.text(98.8, 0.101,'$xbar$ = {}'.format(miu_hat), fontsize=12, color="b")
_= ax.text(106.8, 0.101,'$\mu$ = {}'.format(miu), fontsize=12, color="r")
_= ax.legend()
```
